```python
import numpy as np
import pandas as pd
train=pd.read_csv('housing.csv')
```

```python
train['total_bedrooms']=train['total_bedrooms'].fillna(train['total_bedrooms'].mean())
```

```python
train.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
cat_var =train.dtypes.loc[train.dtypes  == 'object'].index
le =LabelEncoder()
for var in cat_var:
    train[var] = le.fit_transform(train[var])
```

```python
train.head()
```

```python
housing_ind = train.drop("median_house_value",axis=1)
print(housing_ind.head())
housing_dep = train["median_house_value"]
print("Median Hosing Values")
print(housing_dep.head())
```

```python
#splitting
from sklearn import model_selection
X_train,X_test,y_train,y_test = model_selection.train_test_split(housing_ind,housing_dep,test_size=0.2,random_state=42)
print("X_train shape {} and size {}".format(X_train.shape,X_train.size))
print("X_test shape {} and size {}".format(X_test.shape,X_test.size)) 
print("y_train shape {} and size {}".format(y_train.shape,y_train.size))
print("y_test shape {} and size {}".format(y_test.shape,y_test.size))
```

```python
X_train.head()
```

```python
#standardization
from sklearn.preprocessing import StandardScaler
independent_scaler = StandardScaler()
X_train = independent_scaler.fit_transform(X_train)
X_test = independent_scaler.fit_transform(X_test)
print(X_train[0:5,:])
print("test data")
print(X_test[0:5,:])
```

```python
#linear regression
from sklearn.linear_model import LinearRegression
linearRegModel = LinearRegression(n_jobs=-1)
linearRegModel.fit(X_train,y_train)
print("intercept is "+str(linearRegModel.intercept_))
print("coefficients is "+str(linearRegModel.coef_))
```

```python
y_pred=linearRegModel.predict(X_test)
```

```python
print(len(y_pred))
print(len(y_test))
print(y_pred[0:5])
print(y_test[0:5])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
test = pd.DataFrame({'Predicted':y_pred,'Actual':y_test})
fig= plt.figure(figsize=(16,8))
test = test.reset_index()
test =test.drop(['index'],axis=1)
plt.plot(test[:50])
plt.legend(['Actual','Predicted'])

sns.jointplot(x='Actual',y='Predicted',data=test,kind='reg',);
```
